```python
from collections import defaultdict
documents = [
    "Hello world",
    "MapReduce example in Python",
    "Hello Python",
    "Python is awesome",
    "MapReduce is cool",
    "Hello MapReduce"
]
word_counts = defaultdict(int)
for document in documents:
    words = document.split()
    for word in words:
        word_counts[word] += 1
for word, count in word_counts.items():
    print(f"{word}: {count}")
```
